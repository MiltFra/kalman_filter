```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kalman_df = pd.read_csv('cmake-build-debug/kalman_estimates_displacement.csv')
```

```python
kalman_df.index = kalman_df.Time
kalman_df.drop(columns=['Time'])
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)

sns.lineplot(data=[kalman_df.Measurement, kalman_df.Prediction, kalman_df.KalmanEstimate], palette=['b','r','g'])
ax.set(xlabel='Time (s)', ylabel='Displacement (m)', title="Displacement of pod over time")
```
